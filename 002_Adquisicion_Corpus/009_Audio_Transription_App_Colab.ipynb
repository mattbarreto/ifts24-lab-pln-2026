## 2. ImportaciĂ³n de librerĂ­as

Ahora cargamos los componentes necesarios. Vamos a usar el `pipeline` de Transformers, que simplifica enormemente la carga y el uso de modelos pre-entrenados.

```python
import torch
from transformers import pipeline
import gradio as gr

print("LibrerĂ­as cargadas correctamente.")
```

---
## 3. ConfiguraciĂ³n de la lĂ³gica de transcripciĂ³n

Vas a definir una funciĂ³n que reciba la ruta de un archivo de audio y devuelva el texto transcrito. El modelo elegido es `whisper-small`, que ofrece un buen equilibrio entre precisiĂ³n y velocidad de ejecuciĂ³n.

```python
def transcribir_audio(archivo_audio):
    """
    Recibe un archivo de audio y devuelve su transcripciĂ³n usando Whisper.
    """
    # Inicializamos el pipeline de reconocimiento de voz (ASR)
    # El parĂ¡metro chunk_length_s permite procesar audios largos dividiĂ©ndolos en fragmentos
    pipe = pipeline(
        "automatic-speech-recognition",
        model="openai/whisper-small",
        chunk_length_s=30,
        device="cuda" if torch.cuda.is_available() else "cpu"
    )

    # Procesamos el archivo y extraemos el texto
    resultado = pipe(archivo_audio, batch_size=8)["text"]
    return resultado
```

---
## 4. ConstrucciĂ³n de la interfaz grĂ¡fica

Con Gradio, definĂ­s los componentes de entrada (`gr.Audio`) y salida (`gr.Textbox`). La clase `gr.Interface` se encarga de conectar esos componentes con la funciĂ³n que definiste anteriormente.

```python
# Definimos el componente de entrada: audio cargado desde la computadora
entrada_audio = gr.Audio(sources="upload", type="filepath", label="Cargar audio")

# Definimos el componente de salida: cuadro de texto para la transcripciĂ³n
salida_texto = gr.Textbox(label="TranscripciĂ³n resultante")

# Creamos la interfaz conectando lĂ³gica y componentes
demo = gr.Interface(
    fn=transcribir_audio,
    inputs=entrada_audio,
    outputs=salida_texto,
    title="Laboratorio NLP: Transcriptor con Whisper",
    description="SubĂ­ un archivo de audio para que el modelo identifique el idioma y genere la transcripciĂ³n."
)
```
